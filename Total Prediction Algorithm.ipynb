```python
def make_prediction(home,away):
    global hold_dict
    temp_list = [0 for x in range(9)]
    col_names = ['GAME_DATE_EST', 'GAME_ID', 'HOME_TEAM_ID', 'VISITOR_TEAM_ID','SEASON',
                 'PTS_home','FG_PCT_home', 'FT_PCT_home', 'AST_home', 'REB_home','PTS_away',
                 'FG_PCT_away','FT_PCT_away', 'FG3_PCT_away', 'AST_away', 'REB_away']
    
    temp_df = df[col_names].copy()
    
    home_avgs = temp_df.loc[(temp_df['HOME_TEAM_ID']==home) &
                            (temp_df['SEASON'] > 2017)].sort_values(by=['GAME_DATE_EST'],
                            ascending=False).drop_duplicates(subset='GAME_ID').head(3).mean()
    away_avgs = temp_df.loc[(temp_df['VISITOR_TEAM_ID']==away) &
                            (temp_df['SEASON'] > 2017)].sort_values(by=['GAME_DATE_EST'],
                            ascending=False).drop_duplicates(subset='GAME_ID').head(3).mean()
    temp_list = fill_home(temp_list, home_avgs)
    temp_list = fill_away(temp_list, away_avgs)

    
    col_names = ['PTS_home','FG_PCT_home', 'FT_PCT_home', 'AST_home', 'REB_home','PTS_away',
                 'FG_PCT_away','FT_PCT_away', 'FG3_PCT_away', 'AST_away', 'REB_away']
    
    for i in range(len(temp_list)):
        if col_names[i] not in hold_dict:
            hold_dict[col_names[i]] = []
        hold_dict[col_names[i]].append(temp_list[i])
        
def fill_home(temp, avgs):
    avg = avgs[4:]
    order = [0,1,2,3]
    for index in order:
        temp[index] = avg[index]
    return temp

def fill_away(temp, avgs):
    avg = avgs[4:]
    order = [4,5,6,7,8]
    for index in order:
        temp[index] = avg[index]
    return temp
```
