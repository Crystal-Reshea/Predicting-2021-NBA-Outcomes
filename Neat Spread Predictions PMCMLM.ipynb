```python
def make_prediction(home, away):
    global hold_dict
    temp_list = [0 for x in range(27)]
    
    col_names = ['GAME_DATE_EST', 'GAME_ID','Home Team', 'Away Team','FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home','SEASON',
       'AST_home', 'REB_home', 'FG_PCT_away', 'FT_PCT_away',
       'FG3_PCT_away', 'AST_away', 'REB_away', 'HOME_TEAM_WINS',
       'FGA_home', 'OREB_home', 'FGA_away', 'OREB_away', 'TO_diff', 'AST_diff',
       'FTA_diff', 'DREB_diff', 'BLK_home', 'BLK_away', 'BLK_diff','POSS_home','POSS_away', 'WPCT_home', 'WPCT_away']
    temp_df = df[col_names].copy()
    
    home_avgs = temp_df.loc[(df['Home Team']==home) & (df['SEASON'] > 2017)].sort_values(by=['GAME_DATE_EST'], ascending=False).drop_duplicates(subset='GAME_ID').head(3).mean()
#     avgs_home = most_recent_3[['FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home','SEASON',
#        'AST_home', 'REB_home', 'FG_PCT_away', 'FT_PCT_away',
#        'FG3_PCT_away', 'AST_away', 'REB_away', 'HOME_TEAM_WINS',
#        'FGA_home', 'OREB_home', 'FGA_away', 'OREB_away', 'TO_diff', 'AST_diff',
#        'FTA_diff', 'DREB_diff', 'BLK_home', 'BLK_away', 'BLK_diff','POSS_home','POSS_away']]
#     if home == "Houston Rockets":
#         print(temp_df.loc[(df['Home Team']==home) & (df['SEASON'] > 2017)].sort_values(by=['GAME_DATE_EST'], ascending=False).drop_duplicates(subset='GAME_ID').head(3))
        
    away_avgs = temp_df.loc[(df['Home Team']==away) & (df['SEASON'] > 2017)].sort_values(by=['GAME_DATE_EST'], ascending=False).drop_duplicates(subset='GAME_ID').head(3).mean()
#     avgs_home = most_recent_3[['FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home','SEASON',
    
    # Filling in avgerages fof home team
    temp_list = fill_home(temp_list,home_avgs)

    # Filling in averages for away team
    temp_list = fill_away(temp_list, away_avgs)
    
    # Filling in averages for shared points
    temp_list[3] = 2020
    all_avgs = temp_df.loc[((df['Home Team']==home) & (df['Away Team'] ==away)) &(df['SEASON'] > 2017)].sort_values(by=['GAME_DATE_EST'], ascending=False).drop_duplicates(subset='GAME_ID').head(2).mean()
   
    # Reassigning columns 
    col_names = ['FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home','SEASON',
       'AST_home', 'REB_home', 'FG_PCT_away', 'FT_PCT_away',
       'FG3_PCT_away', 'AST_away', 'REB_away', 'HOME_TEAM_WINS',
       'FGA_home', 'OREB_home', 'FGA_away', 'OREB_away', 'TO_diff', 'AST_diff',
       'FTA_diff', 'DREB_diff', 'BLK_home', 'BLK_away', 'BLK_diff','POSS_home','POSS_away','WPCT_home', 'WPCT_away']

    for i in range(len(temp_list)):
        if col_names[i] not in hold_dict:
            hold_dict[col_names[i]] = []
        hold_dict[col_names[i]].append(temp_list[i])

        
        
def fill_home(temp, avgs):
    order = [0,1,2,4,5,12,13,20,23,25]
    avg = avgs[1:]
    for index in order:
        temp[index] = avg[index]
    return temp

def fill_away(temp, avgs):
    order =[6,7,8,9,10,14,15,21,24,26]
    avg = avgs[1:]
    for index in order:
        temp[index] = avg[index]
    return temp

def fill_both(temp, avgs):
    order = [11,16,17,18,19,22]
    avg = avgs[1:]
    for index in order:
        temp[index] = avg[index]
    return temp
```
